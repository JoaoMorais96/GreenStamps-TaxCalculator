```python
# define bracket we want
bracket = 2

# get specific percentage 
perc = df.loc(axis=1)['Número de dependentes','0'].iloc[bracket]
print(float(perc.replace("%", "").replace(",", "."))/100)

# get specific salary
salary = df.loc(axis=1)['Remuneração Mensal Euros','Valor Bruto'].iloc[bracket]
print(salary)
```

## But now we want the values according to our salary, without knowing the specific bracket

```python
from bisect import bisect
def getBracketAndTax(df, salary):
    max_sals = df.loc(axis=1)['Remuneração Mensal Euros','Valor Bruto']
    bracket = bisect(max_sals, salary)

    if (bracket == len(max_sals)):
        upperLim = 'infinite'
        bracket=bracket-1
    else:
        upperLim = df.loc(axis=1)['Remuneração Mensal Euros','Valor Bruto'].iloc[bracket]

    if(bracket==0):
        lowerLim = 0       
    else:
        lowerLim = df.loc(axis=1)['Remuneração Mensal Euros','Valor Bruto'].iloc[bracket-1]
        bracket=bracket-1
    

    print('A taxable salary of', salary,'euros is in the bracket',bracket, 'because it is higher than', lowerLim,
          'and lower than', upperLim,'.')
    
    perc = df.loc(axis=1)['Número de dependentes','0'].iloc[bracket]
    
    print('It will pay',perc,'in tax.\n')
    return float(perc.replace("%", "").replace(",", "."))/100

getBracketAndTax(df, 630)
getBracketAndTax(df, 770)
getBracketAndTax(df, 1450)
getBracketAndTax(df, 27014.0)
print('End test')
```

## Now we need a function to calculate the amount of tax that will be paid

```python
def netVal(df, salary, socialSec, nonTaxable):
    # get non taxable salary
    noTax = salary * nonTaxable
    ogSalary = salary
    salary = salary - noTax

    # get tax percentage for the taxable income
    taxPerc = getBracketAndTax(df, salary)

    # calculate tax (social security+tax)
    taxed = salary*(socialSec) + salary*(taxPerc)

    print('So if you are using green stamps in Portugal, a gross salary off',ogSalary,
          'euros results in a net salary of',round((ogSalary-taxed),2),'euros.')
    print('\nSummary:\n Gross Salary:',ogSalary,'\n Taxable Salary:', salary, '\n Net Salary:', round((ogSalary-taxed),2),
        '\n Monthly Tax:', round((ogSalary - (ogSalary-taxed)),2), '\n Annual Tax:', round(12*(ogSalary - (ogSalary-taxed)),2))
```

```python
grossSal = 8334
socialsec = 20
nonTaxable = 30 

socialsec = float(socialsec/100)
nonTaxable = float(nonTaxable/100)

netVal(df, grossSal, socialsec, nonTaxable)
```
